**In this notebook we'll run some model tests and look at a base model to get an initial idea of our model summary and perform inital feature selection/filtering.**

```python
import pandas as pd
import pickle as pkl
from module2_scripts.model import ModelTests, BaseModel
```

```python
X = pd.read_csv('working_datasets/df_preprocessed_X.csv')
y = pd.read_csv('working_datasets/df_preprocessed_y.csv') 
```

```python
#initialize ModelTests instance object
obj = ModelTests(X, y)
```

```python
#initialize .run_tests() method to display OLS reports for each variable
obj.run_tests()
```

The above cell executed calls the run_tests() method on the MotelTests instance object and returns the OLS regresseion summary report for each variable, and calculated and displays the sum of squares for each model run, ordered from least to greatest. We generally want to keep the variables that have lower SSE. For categorical variables that have been one-hot encoded, the method groups all the dummy variables together and then runs the model, you can see the 'view' and 'zipcode' OLS summaries for reference. 

Individually, none of the variables have strong R-squared figures. But, we expect the overall base model R-squared to increase once all the variables are grouped together and then the model is ran, and then we expect R-squared to increase further once we do further feature selection and introduce interaction variables into the model. 

For now, we will focus on again filtering the features manually, considering which features would it make sense to leave in the model and which the model would be better off without, and also considering the SSE for each variable. 

To start, focusing our attention on the latitude and longitude variable pairs, these coordinates do not work very well with the model. These have not been log transformed as these are moreso descriptor variables rather than continuous variables. Also, the zipcode variable and the lat-long variable pairs more or less tell us similar info, which is info regarding the geographic location of the unit. Since zipcode ranked #3 in lowest SSE, it makes sense to keep the zipcode variables and drop the **lat and long** variable pairs. 

Another clear variable to drop is the **date_delta** variable. This is a derivative of the date variable. Since the statsmodels.api OLS function does not work with date objects, this variable had to be converted to the n number of days since the earliest date in the set. 


```python
#initialize BaseModel object
obj_base = BaseModel(X, y)
```

```python
obj.ols_summary()
```

Running the OLS regression for the base model, we see that adjusted R-squared is 69.5%! Overall, the tests are okay given the nature of the dataset and the fact that the dependent variable has not been log transformed, however, the report gives us a warning that the condition number is large, which might indicate strong multicollinearity or "other numerical problems". Let's drop the aforementioned variables and rerun the base model and note the changes. 

```python
X_processed = X.drop(['lat', 'long', 'date_delta'], axis=1)
```

```python
obj_base_processed = BaseModel(X_processed, y)
obj_base_processed.ols_summary()
```

We see that our updated OLS report reflects a very minor decrease in adjusted R-squared to 69.1%. A decrease is expected since we removed some variables from the model. The more variables that are included in the model, the more R-squared artificially increases. The good news is that we dropped 3 variables and the R-squared only re-adjusted -0.4%, which shows that the remaining variables still hold a lot of value to the model. We also note that the condition number is very low and the multicollinearity and other numberical problems warning is now gone. 

Now, we're ready to proceed to the next step of the model, which is to do further feature selection using stepwise selection method, and add interaction variables for better model fit.
